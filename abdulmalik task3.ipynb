## peak hours of sale

```python
df['hour'] = pd.to_datetime(df['time']).dt.hour
peak_hours = df.groupby('hour')['quantity'].sum().sort_values(ascending=False)
peak_hours.head()
plt.bar(peak_hours.index, peak_hours.values, color='orange',)
plt.title("Sales by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Quantity Sold")
plt.show()

```

## sales made on each day of the week 

```python
df['date'] = pd.to_datetime(df['date'])
df['day_of_week'] = df['date'].dt.day_name()

sales_by_day = df.groupby('day_of_week')['quantity'].sum().sort_values(ascending=False)
plt.bar(sales_by_day.index, sales_by_day.values, color='skyblue',)
plt.title("Total Sales by Day of Week")
plt.ylabel("Quantity Sold") 
plt.show()
```

## Top 5 best selling pizzas

```python
top_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=False).head(5)
top_pizzas.plot(kind='bar', color='green')
plt.title("Top 5 Bestselling Pizzas")
plt.ylabel("Quantity Sold")
plt.show()
```
